# Przykładowe użycie Jupytera

Najpierw trzeba zimportować paczki z których będziemy korzystać.

```julia
using Printf
using PyPlot
```

### Wypisywanie @printf
Do wypisywania danych można użyć printfa tak jak w C.

```julia
x = 100
y = 3.14
z = "jakiś string"
@printf("Integer: %d\nFloat:%f\nString:%s\n", x, y, z)
```

### Rysowanie wykresów
Wykresy można rysować przy pomocy biblioteki PyPlot

```julia
xrange = 1:100
values = rand(100)
plot(xrange, values, linewidth=2.0, linestyle="-")
```
